### 信息熵

```python
# -*- coding:utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.rcParams['axes.unicode_minus']=False

def entropy(p):
    return -p*np.log(p)-(1-p)*np.log(1-p)

p=np.linspace(0,1,200,endpoint=False)[1:]

entropy_list=entropy(p)

plt.plot(p,entropy_list)
plt.xlabel(u'概率值')
plt.ylabel(u'系统信息熵')
plt.title(u'二分类问题系统信息熵随概率变化')
plt.legend()
plt.show()
```

### 基尼系数

```python
# -*- coding:utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.rcParams['axes.unicode_minus']=False

def gini(p):
    return 1-p**2-(1-p)**2

p=np.linspace(0,1,200,endpoint=False)[1:]

entropy_list=gini(p)

plt.plot(p,entropy_list)
plt.xlabel(u'概率值')
plt.ylabel(u'系统基尼系数')
plt.title(u'二分类问题系统激励不纯度随概率变化')
plt.legend()
plt.show()
```
