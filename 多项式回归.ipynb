### 多项式回归：对样本数据进行升维，然后在高维空间中使用线性回归的思路进行回归拟合

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import  matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x=np.random.uniform(-3,3,size=100)
X=x.reshape(-1,1)
y=0.8*x**2+x+2+np.random.normal(0,1,100)
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
plt.scatter(x,y)
plt.show()
```

### 使用Pipeline组合各个流程

```python
poly_reg_pipe=Pipeline([
    ("poly",PolynomialFeatures(degree=2)),
    ('std',StandardScaler()),
    ('lin_reg',LinearRegression())])

poly_reg_pipe.fit(X_train,y_train)

print poly_reg_pipe.score(X_test,y_test)

y_predict=poly_reg_pipe.predict(X)

plt.scatter(x,y)
plt.plot(np.sort(x),y_predict[np.argsort(x)],color='r')
plt.show()
```

```python
poly_reg_pipe=Pipeline([
    ("poly",PolynomialFeatures(degree=1)),
    ('std',StandardScaler()),
    ('lin_reg',LinearRegression())])

poly_reg_pipe.fit(X_train,y_train)

print poly_reg_pipe.score(X_test,y_test)

y_predict=poly_reg_pipe.predict(X)

plt.scatter(x,y)
plt.plot(np.sort(x),y_predict[np.argsort(x)],color='r')
plt.show()
```
