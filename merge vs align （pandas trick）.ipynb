### merge 和 align 的使用
note：对类别类变量进行热编码之后，需保证test_data 和 train_data 特征顺序和保持完全的一致性，这时可能需要用到align操作。 

```python
df1.merge(df2, how='outer', right_index=True, left_index=True)
```

```python
x,y=df1.align(df2,join='left',axis=1)
```
